```python
class TrafficDevice:
    def activate(self):
        print("Activating generic traffic device...")


class TrafficLight(TrafficDevice):
    def activate(self):
        print("Traffic Light: Changing signal state (Red -> Yellow -> Green).")

class SpeedCamera(TrafficDevice):
    def activate(self):
        print("Speed Camera: Monitoring vehicle speeds and capturing license plates.")

class PedestrianSignal(TrafficDevice):
    def activate(self):
        print("Pedestrian Signal: Displaying 'WALK' sign and enabling audio cue.")

class EmergencySiren(TrafficDevice):
    def activate(self):
        print("Emergency Siren: Sounding alarm and flashing warning lights.")

def main():
    # Store objects in a list
    devices = [
        TrafficLight(),
        SpeedCamera(),
        PedestrianSignal(),
        EmergencySiren()
    ]

    print("--- Activating Smart Traffic Management System ---")
    for device in devices:
        device.activate()

if __name__ == "__main__":
    main()
```
